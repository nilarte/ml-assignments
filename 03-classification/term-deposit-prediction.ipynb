```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_full = pd.read_csv('bank-full.csv')
df_full
```

```python
selected_columns = [
    'age', 'job', 'marital', 'education', 'balance', 
    'housing', 'contact', 'day', 'month', 'duration', 
    'campaign', 'pdays', 'previous', 'poutcome', 'y'
]

# Select the desired columns
df = df_full[selected_columns]
df
```

```python
df.isnull().sum()
```

```python
# Find the mode (most frequent value) for the 'education' column
education_mode = df['education'].mode()[0]
print(f"\nMost frequent observation (mode) in 'education': {education_mode}")
```

```python
import pandas as pd

# Select only the numerical features from the DataFrame
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Display the correlation matrix
print("Correlation Matrix:\n", correlation_matrix)

# Find the two features with the highest correlation (excluding self-correlation)
correlation_matrix_unstacked = correlation_matrix.unstack()
sorted_correlations = correlation_matrix_unstacked.sort_values(ascending=False)

# Exclude self-correlations (where feature correlates with itself)
top_correlated_pair = sorted_correlations[sorted_correlations < 1].idxmax()
highest_correlation = sorted_correlations[sorted_correlations < 1].max()

print(f"\nThe two most correlated features are: {top_correlated_pair}")
print(f"With a correlation coefficient of: {highest_correlation}")
```

```python
df.y = (df.y == 'yes').astype(int)
```

```python
from sklearn.model_selection import train_test_split
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
y_train = df_train.y.values
y_val = df_val.y.values
y_test = df_test.y.values

del df_train['y']
del df_val['y']
del df_test['y']
```

```python
from sklearn.metrics import mutual_info_score
mutual_info_score(y_train, df_train.contact).round(2)

```

```python
mutual_info_score(y_train, df_train.education).round(2)
```

```python
mutual_info_score(y_train, df_train.housing).round(2)
```

```python
mutual_info_score(y_train, df_train.poutcome).round(2)
```
